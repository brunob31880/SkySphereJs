```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def draw_modified_declination_line(ax, M):
    """Dessiner une ligne en pointillé de M à (5,5,0) passant par (6,3,6) et ajouter le texte "Déclinaison"."""
    I = np.array([3, 6, 6])  # Point intermédiaire (6,3,6)
    J = np.array([0, 5, 5])  # Coordonnées du point où M se projette sur l'axe Z
    
    # Tracer les segments M-I et I-J
    ax.plot([M[1], I[1]], [M[2], I[2]], [M[0], I[0]], linestyle='--', color='k')
    ax.plot([I[1], J[1]], [I[2], J[2]], [I[0], J[0]], linestyle='--', color='k')
    
    ax.text((M[1] + I[1])/2, (M[2] + I[2])/2, (M[0] + I[0])/2, 'Déclinaison', color='black', fontsize=10)
    
    
def draw_modified_right_ascension_line_v2(ax, H):
    """Dessiner une ligne en pointillé de H à (5,0,0) passant par (6,0,2) et ajouter le texte "Ascension droite"."""
    I = np.array([0, 2, 6])  # Point intermédiaire (6,0,2)
    J = np.array([0, 0, 5])  # Coordonnées du point où H se projette sur l'axe Z
    
    # Tracer les segments H-I et I-J
    ax.plot([H[1], I[1]], [H[2], I[2]], [H[0], I[0]], linestyle='--', color='k')
    ax.plot([I[1], J[1]], [I[2], J[2]], [I[0], J[0]], linestyle='--', color='k')
    
    ax.text((H[1] + I[1])/2, (H[2] + I[2])/2, (H[0] + I[0])/2, 'Ascension droite', color='black', fontsize=10)

# Créer la figure avec le repère déjà défini
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Dessiner les axes
ax.quiver(0, 0, 0, 10, 0, 0, color='b', arrow_length_ratio=0.05)
ax.quiver(0, 0, 0, 0, 10, 0, color='r', arrow_length_ratio=0.05)
ax.quiver(0, 0, 0, 0, 0, 10, color='g', arrow_length_ratio=0.05)

# Ajouter le point M
M = np.array([5, 5, 5])

# Coordonnées du point H
H = np.array([0, 5, 5])

# Ajouter les points M et H
ax.scatter(M[1], M[2], M[0], c='black', s=100, label='M(5,5,5)')
ax.scatter(H[1], H[2], H[0], c='blue', s=100, label='H(5,5,0)')

# Tracer la ligne de déclinaison et la ligne d'ascension droite modifiée
draw_modified_declination_line(ax, M)
draw_modified_right_ascension_line_v2(ax, H)

# Tracer les rayons OM et OH avec des flèches à leurs extrémités
ax.quiver(0, 0, 0, M[1], M[2], M[0], color='purple', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, H[1], H[2], H[0], color='cyan', arrow_length_ratio=0.1)

# Configurer les axes, la vue et les labels
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('Z (Bleu)')
ax.set_ylabel('X (Rouge)')
ax.set_zlabel('Y (Vert)')
ax.view_init(elev=20, azim=-45)
ax.legend()

plt.show()
```
